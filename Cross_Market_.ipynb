11.Find the highest oil price in the last 5 years.


```python
queries["11. Highest oil price (5 years)"] = """
SELECT strftime('%Y-%m-%d', Date) as date, MAX(Price) as highest_price
FROM oil_prices
WHERE Date >= date('now', '-5 years')
GROUP BY strftime('%Y-%m-%d', Date)
ORDER BY highest_price DESC
LIMIT 1;"""
```

12.Get the average oil price per year.


```python
queries["12. Avg oil price per year"] = """
SELECT strftime('%Y', Date) as year,
       ROUND(AVG(Price), 2) as avg_price
FROM oil_prices
GROUP BY year
ORDER BY year DESC;"""
```

13.Show oil prices during COVID crash (March–April 2020).


```python
queries["13. Oil prices during COVID crash"] = """
SELECT strftime('%Y-%m-%d', Date) as date, Price,
       LAG(Price) OVER (ORDER BY Date) as prev_price,
       ROUND(Price - LAG(Price) OVER (ORDER BY Date), 2) as daily_change
FROM oil_prices
WHERE Date >= '2020-03-01' AND Date <= '2020-04-30'
ORDER BY Date;"""
```

14.Find the lowest price of oil in the last 10 years.

7.Calculate the average daily price of Ethereum in the past 1 year.


```python
queries["7. Ethereum average price (1 year)"] = """
SELECT coin_id,
       ROUND(AVG(price_usd), 2) as avg_price,
       MIN(price_usd) as min_price,
       MAX(price_usd) as max_price
FROM crypto_prices
WHERE coin_id = 'ethereum'
  AND date >= date('now', '-1 year')
GROUP BY coin_id;"""
```

8.Show the daily price trend of Bitcoin in june 2026 .

```python
queries["8. Bitcoin daily price trend (jun 2026)"] = """
SELECT date,
       price_usd,
       LAG(price_usd) OVER (ORDER BY date) as prev_price,
       ROUND(price_usd - LAG(price_usd) OVER (ORDER BY date), 2) as price_change
FROM crypto_prices
WHERE coin_id = 'bitcoin'
  AND date >= '2025-12-01'
  AND date < '2026-06-01'
ORDER BY date;"""
```

10.Get the % change in Bitcoin’s price between jan 2026 and jun 2026


```python
queries["10. Bitcoin % change ( jan 2020 vs jun 2026)"] = """
SELECT
    MAX(CASE WHEN date >= '2026-01-01' AND date < '2026-06-01' THEN price_usd END) as jan_2026,
    MAX(CASE WHEN date >= '2026-01-01' AND date < '2026-06-01' THEN price_usd END) as jun_2026,
    ROUND(((MAX(CASE WHEN date >= '2026-01-01' AND date < '2026-06-01' THEN price_usd END) -
            MAX(CASE WHEN date >= '2026-01-01' AND date < '2026-06-01' THEN price_usd END)) * 100.0 /
            NULLIF(MAX(CASE WHEN date >= '2026-01-01' AND date < '2026-06-01' THEN price_usd END), 0)), 2) as pct_change
FROM crypto_prices
WHERE coin_id = 'bitcoin';"""
```

29.Join top 3 crypto coins with stock indices for 2026.


```python
queries["29. Join top 3 crypto coins with stock indices for 2026."] = """
SELECT
    cp.date,
    MAX(CASE WHEN cp.coin_id = 'bitcoin' THEN cp.price_usd ELSE NULL END) AS bitcoin_price,
    MAX(CASE WHEN cp.coin_id = 'ethereum' THEN cp.price_usd ELSE NULL END) AS ethereum_price,
    MAX(CASE WHEN cp.coin_id = 'tether' THEN cp.price_usd ELSE NULL END) AS tether_price,
    MAX(CASE WHEN s.Ticker = '^GSPC' THEN s.Close ELSE NULL END) AS gspc_close_price,
    MAX(CASE WHEN s.Ticker = '^IXIC' THEN s.Close ELSE NULL END) AS ixic_close_price,
    MAX(CASE WHEN s.Ticker = '^NSEI' THEN s.Close ELSE NULL END) AS nsei_close_price
FROM crypto_prices cp
JOIN stocks s ON cp.date = DATE(s.Date)
WHERE strftime('%Y', cp.date) = '2026'
  AND cp.coin_id IN ('bitcoin', 'ethereum', 'tether')
  AND s.Ticker IN ('^GSPC', '^IXIC', '^NSEI')
GROUP BY cp.date
ORDER BY cp.date DESC
LIMIT 30;"""
```

```python
%%writefile app.py

import streamlit as st
import pandas as pd
import sqlite3
from datetime import datetime

# Define SQL_QUERIES dictionary
SQL_QUERIES = {
    "1. Top 3 cryptocurrencies by market cap": """
SELECT name, symbol, market_cap, market_cap_rank, current_price
FROM cryptocurrencies
ORDER BY market_cap_rank
LIMIT 3;""",
    "2. Coins with >90% circulating supply": """
SELECT name, symbol, circulating_supply, total_supply,
       ROUND((circulating_supply*100.0/NULLIF(total_supply, 0)),2) as supply_percentage
FROM cryptocurrencies
WHERE total_supply > 0
  AND (circulating_supply*100.0/total_supply) > 90
ORDER BY supply_percentage DESC;""",
    "3. Coins within 10% of ATH": """
SELECT name, symbol, current_price, ath,
       ROUND((current_price*100.0/ath),2) as percent_of_ath
FROM cryptocurrencies
WHERE ath > 0
  AND (current_price*100.0/ath) >= 90
ORDER BY percent_of_ath DESC;""",
    "4. Avg market cap rank (volume > $1B)": """
SELECT ROUND(AVG(market_cap_rank), 2) as avg_rank,
       COUNT(*) as coin_count
FROM cryptocurrencies
WHERE total_volume > 1000000000;""",
    "5. Most recently updated coin": """
SELECT name, symbol, current_price, market_cap, last_updated
FROM cryptocurrencies
ORDER BY last_updated DESC
LIMIT 1;""",
    "6. Highest Bitcoin price (last 365 days)": """
SELECT date, MAX(price_usd) as highest_price
FROM crypto_prices
WHERE coin_id = 'bitcoin'
  AND date >= date('now', '-365 days')
GROUP BY date
ORDER BY highest_price DESC
LIMIT 1;""",
    "7. Ethereum average price (1 year)": """
SELECT coin_id,
       ROUND(AVG(price_usd), 2) as avg_price,
       MIN(price_usd) as min_price,
       MAX(price_usd) as max_price
FROM crypto_prices
WHERE coin_id = 'ethereum'
  AND date >= date('now', '-1 year')
GROUP BY coin_id;""",
    "8. Bitcoin daily price trend (jun 2026)": """
SELECT date,
       price_usd,
       LAG(price_usd) OVER (ORDER BY date) as prev_price,
       ROUND(price_usd - LAG(price_usd) OVER (ORDER BY date), 2) as price_change
FROM crypto_prices
WHERE coin_id = 'bitcoin'
  AND date >= '2025-12-01'
  AND date < '2026-06-01'
ORDER BY date;""",
    "9. Coin with highest avg price (1 year)": """
SELECT coin_id,
       ROUND(AVG(price_usd), 2) as avg_price
FROM crypto_prices
WHERE date >= date('now', '-1 year')
GROUP BY coin_id
ORDER BY avg_price DESC
LIMIT 1;""",
    "10. Bitcoin % change ( jan 2020 vs jun 2026)": """
SELECT
    MAX(CASE WHEN date >= '2026-01-01' AND date < '2026-06-01' THEN price_usd END) as jan_2026,
    MAX(CASE WHEN date >= '2026-01-01' AND date < '2026-06-01' THEN price_usd END) as jun_2026,
    ROUND(((MAX(CASE WHEN date >= '2026-01-01' AND date < '2026-06-01' THEN price_usd END) -
            MAX(CASE WHEN date >= '2026-01-01' AND date < '2026-06-01' THEN price_usd END)) * 100.0 /
            NULLIF(MAX(CASE WHEN date >= '2026-01-01' AND date < '2026-06-01' THEN price_usd END), 0)), 2) as pct_change
FROM crypto_prices
WHERE coin_id = 'bitcoin';""",
    "11. Highest oil price (5 years)": """
SELECT strftime('%Y-%m-%d', Date) as date, MAX(Price) as highest_price
FROM oil
WHERE Date >= date('now', '-5 years')
GROUP BY date
ORDER BY highest_price DESC
LIMIT 1;""",
    "12. Avg oil price per year": """
SELECT strftime('%Y', Date) as year,
       ROUND(AVG(Price), 2) as avg_price
FROM oil
GROUP BY year
ORDER BY year DESC;""",
    "13. Oil prices during COVID crash": """
SELECT strftime('%Y-%m-%d', Date) as date, Price,
       LAG(Price) OVER (ORDER BY Date) as prev_price,
       ROUND(Price - LAG(Price) OVER (ORDER BY Date), 2) as daily_change
FROM oil
WHERE Date >= '2020-03-01' AND Date <= '2020-04-30'
ORDER BY Date;""",
    "14. Lowest oil price (10 years)": """
SELECT strftime('%Y-%m-%d', Date) as date, MIN(Price) as lowest_price
FROM oil
WHERE Date >= date('now', '-10 years')
GROUP BY date
ORDER BY lowest_price ASC
LIMIT 1;""",
    "15. Oil price volatility per year": """
SELECT strftime('%Y', Date) as year,
       MAX(Price) - MIN(Price) as volatility
FROM oil
GROUP BY year
ORDER BY volatility DESC;""",
    "16. Get all stock prices for a given ticker": """
SELECT Date, Open, High, Low, Close, Volume
FROM stocks
WHERE Ticker = '{{ticker_symbol}}'
ORDER BY Date;""",
    "17. Find the highest closing price for NASDAQ (^IXIC)": """
SELECT MAX(Close) as highest_closing_price
FROM stocks
WHERE Ticker = '^IXIC';""",
    "18. List top 5 days with highest price difference (high - low) for S&P 500 (^GSPC)": """
SELECT Date, High, Low, ROUND(High - Low, 2) AS price_difference
FROM stocks
WHERE Ticker = '^GSPC'
ORDER BY price_difference DESC
LIMIT 5;""",
    "19. Get monthly average closing price for each ticker": """
SELECT Ticker, strftime('%Y-%m', Date) as year_month,
       ROUND(AVG(Close), 2) as average_closing_price
FROM stocks
GROUP BY Ticker, year_month
ORDER BY Ticker, year_month;""",
    "20. Get average trading volume of NSEI in 2024": """
SELECT ROUND(AVG(Volume), 2) AS average_trading_volume
FROM stocks
WHERE Ticker = '^NSEI' AND strftime('%Y', Date) = '2024';""",
    "21. Compare Bitcoin vs Oil average price in 2026": """
SELECT
  (SELECT ROUND(AVG(cp.price_usd), 2)
   FROM crypto_prices cp
   WHERE cp.coin_id = 'bitcoin' AND strftime('%Y', cp.date) = '2026') AS avg_bitcoin_price_2026,
  (SELECT ROUND(AVG(o.Price), 2)
   FROM oil o
   WHERE strftime('%Y', o.Date) = '2026') AS avg_oil_price_2026;""",
    "22. Check if Bitcoin moves with S&P 500 (correlation idea)": """
SELECT cp.date, cp.price_usd as bitcoin, s.Close as sp500
FROM crypto_prices cp
JOIN stocks s ON cp.date = DATE(s.Date)
WHERE cp.coin_id = 'bitcoin'
  AND s.Ticker = '^GSPC'
ORDER BY cp.date DESC
LIMIT 30;""",
    "23. Compare Ethereum and NASDAQ daily prices for 2026": """
SELECT cp.date, cp.price_usd as ethereum, s.Close as nasdaq
FROM crypto_prices cp
JOIN stocks s ON cp.date = DATE(s.Date)
WHERE cp.coin_id = 'ethereum'
  AND s.Ticker = '^IXIC'
ORDER BY cp.date DESC
LIMIT 30;""",
    "24. Find days when oil price spiked and compare with Bitcoin price change": """
SELECT DATE(op.Date) as date, op.Price as oil, cp.price_usd as bitcoin
FROM oil op
JOIN crypto_prices cp ON DATE(op.Date) = cp.date
WHERE cp.coin_id = 'bitcoin'
ORDER BY op.Date DESC
LIMIT 30;""",
    "25. Compare top 3 coins daily price trend vs Nifty (^NSEI)": """
SELECT
    cp.date,
    s.Close AS nifty_close_price,
    MAX(CASE WHEN cp.coin_id = 'bitcoin' THEN cp.price_usd ELSE NULL END) AS bitcoin_price,
    MAX(CASE WHEN cp.coin_id = 'ethereum' THEN cp.price_usd ELSE NULL END) AS ethereum_price,
    MAX(CASE WHEN cp.coin_id = 'tether' THEN cp.price_usd ELSE NULL END) AS tether_price
FROM crypto_prices cp
JOIN stocks s ON cp.date = DATE(s.Date)
WHERE s.Ticker = '^NSEI'
  AND cp.coin_id IN ('bitcoin', 'ethereum', 'tether')
GROUP BY cp.date, s.Close
ORDER BY cp.date DESC
LIMIT 30;""",
    "26. Compare stock prices (^GSPC) with crude oil prices on the same dates": """
SELECT
    s.Date,
    s.Close AS gspc_close_price,
    o.Price AS oil_price
FROM stocks s
JOIN oil o ON s.Date = o.Date
WHERE s.Ticker = '^GSPC'
ORDER BY s.Date DESC
LIMIT 30;""",
    "27. Correlate Bitcoin closing price with crude oil closing price (same date)": """
SELECT
    cp.date,
    cp.price_usd AS bitcoin_price,
    o.Price AS oil_price
FROM crypto_prices cp
JOIN oil o ON cp.date = DATE(o.Date)
WHERE cp.coin_id = 'bitcoin'
ORDER BY cp.date DESC
LIMIT 30;""",
    "28. Compare NASDAQ (^IXIC) with Ethereum price trends": """
SELECT
    cp.date,
    cp.price_usd AS ethereum_price,
    s.Close AS nasdaq_close_price
FROM crypto_prices cp
JOIN stocks s ON cp.date = DATE(s.Date)
WHERE cp.coin_id = 'ethereum'
  AND s.Ticker = '^IXIC'
ORDER BY cp.date DESC
LIMIT 30;""",
    "29. Join top 3 crypto coins with stock indices for 2026.": """
SELECT
    cp.date,
    MAX(CASE WHEN cp.coin_id = 'bitcoin' THEN cp.price_usd ELSE NULL END) AS bitcoin_price,
    MAX(CASE WHEN cp.coin_id = 'ethereum' THEN cp.price_usd ELSE NULL END) AS ethereum_price,
    MAX(CASE WHEN cp.coin_id = 'tether' THEN cp.price_usd ELSE NULL END) AS tether_price,
    MAX(CASE WHEN s.Ticker = '^GSPC' THEN s.Close ELSE NULL END) AS gspc_close_price,
    MAX(CASE WHEN s.Ticker = '^IXIC' THEN s.Close ELSE NULL END) AS ixic_close_price,
    MAX(CASE WHEN s.Ticker = '^NSEI' THEN s.Close ELSE NULL END) AS nsei_close_price
FROM crypto_prices cp
JOIN stocks s ON cp.date = DATE(s.Date)
WHERE strftime('%Y', cp.date) = '2026'
  AND cp.coin_id IN ('bitcoin', 'ethereum', 'tether')
  AND s.Ticker IN ('^GSPC', '^IXIC', '^NSEI')
GROUP BY cp.date
ORDER BY cp.date DESC
LIMIT 30;""",
    "30. Multi-join: stock prices, oil prices, and Bitcoin prices for daily comparison": """
SELECT
    cp.date,
    cp.price_usd AS bitcoin_price,
    o.Price AS oil_price,
    s.Close AS gspc_close_price
FROM crypto_prices cp
JOIN oil o ON cp.date = DATE(o.Date)
JOIN stocks s ON cp.date = DATE(s.Date)
WHERE cp.coin_id = 'bitcoin'
  AND s.Ticker = '^GSPC'
ORDER BY cp.date DESC
LIMIT 30;"""
}

def get_conn():
    return sqlite3.connect("mydb.db")

st.set_page_config(page_title="Cross-Market Analytics", layout="wide", page_icon="📈")

st.title("💰🛢️📈 Cross-Market Analysis: Crypto, Oil & Stocks")
st.caption("Bitcoin • Ethereum • Tether | WTI Crude Oil | S&P 500 • NASDAQ • NIFTY")

# Verify DB
conn = get_conn()
try:
    count = pd.read_sql("SELECT COUNT(*) as n FROM crypto_prices", conn).iloc[0]["n"]
    st.sidebar.success(f"✅ DB connected — {count} crypto price rows")
except:
    st.error("❌ market.db not found or empty. Run the INSERT cell first!")
    st.stop()
conn.close()

page = st.sidebar.radio("📄 Navigate to", [
    "📊 Page 1: Market Overview",
    "🧠 Page 2: SQL Query Runner",
    "🪙 Page 3: Top 3 Crypto Analysis"
])

# ─────────────────────────────────────────────
# PAGE 1
# ─────────────────────────────────────────────
if page == "📊 Page 1: Market Overview":
    st.header("🔍 Market Overview")

    col1, col2 = st.columns(2)
    with col1:
        start_date = st.date_input("📅 Start Date", value=datetime(2025, 1, 1))
    with col2:
        end_date = st.date_input("📅 End Date", value=datetime(2026, 6, 9))

    s, e = str(start_date), str(end_date)
    conn = get_conn()

    # ── KPI Cards ──
    st.subheader("📌 Average Prices in Selected Period")
    k1, k2, k3, k4 = st.columns(4)

    try:
        val = pd.read_sql(f"SELECT ROUND(AVG(price_usd),2) as v FROM crypto_prices WHERE coin_id='bitcoin' AND date BETWEEN '{s}' AND '{e}'", conn).iloc[0]["v"]
        k1.metric("₿ Bitcoin (USD)", f"${val:, .0f}" if val else "N/A")
    except: k1.metric("Bitcoin", "N/A")

    try:
        val = pd.read_sql(f"SELECT ROUND(AVG(Price),2) as v FROM oil WHERE date BETWEEN '{s}' AND '{e}'", conn).iloc[0]["v"]
        k2.metric("🛢️ Oil (USD/barrel)", f"${val:.2f}" if val else "N/A")
    except: k2.metric("Oil", "N/A")

    try:
        val = pd.read_sql(f"SELECT ROUND(AVG(Close),2) as v FROM stocks WHERE Ticker='^GSPC' AND Date BETWEEN '{s}' AND '{e}'", conn).iloc[0]["v"]
        k3.metric("📊 S&P 500 Close", f"${val:,.2f}" if val else "N/A")
    except: k3.metric("S&P 500", "N/A")

    try:
        val = pd.read_sql(f"SELECT ROUND(AVG(Close),2) as v FROM stocks WHERE Ticker='^NSEI' AND Date BETWEEN '{s}' AND '{e}'", conn).iloc[0]["v"]
        k4.metric("📉 NIFTY Close", f"₹{val:,.2f}" if val else "N/A")
    except: k4.metric("NIFTY", "N/A")

    # ── Daily Snapshot Table ──
    st.subheader("📋 Daily Market Snapshot (BTC + Oil + S&P 500 + NIFTY)")
    try:
        snap = pd.read_sql(f"""
            SELECT cp.date,
                   ROUND(cp.price_usd,2) as btc_usd,
                   ROUND(o.Price,2)  as oil_usd,
                   ROUND(sp.Close,2)     as sp500_close,
                   ROUND(ns.Close,2)     as nifty_close
            FROM crypto_prices cp
            JOIN oil o  ON cp.date = DATE(o.Date)
            JOIN stocks sp ON cp.date = DATE(sp.Date) AND sp.Ticker='^GSPC'
            JOIN stocks ns ON cp.date = DATE(ns.Date) AND ns.Ticker='^NSEI'
            WHERE cp.coin_id='bitcoin'
              AND cp.date BETWEEN '{s}' AND '{e}'
            ORDER BY cp.date DESC
            LIMIT 200
        """, conn)
        st.dataframe(snap, use_container_width=True, height=400)

        # ── Line charts ──
        if not snap.empty:
            st.subheader("📈 Bitcoin Price Trend")
            st.line_chart(snap.set_index("date")["btc_usd"])

            st.subheader("🛢️ Oil Price Trend")
            st.line_chart(snap.set_index("date")["oil_usd"])

            st.subheader("📊 S&P 500 vs NIFTY")
            st.line_chart(snap.set_index("date")[["sp500_close","nifty_close"]])
    except Exception as ex:
        st.warning(f"Snapshot query: {ex}")

    conn.close()

# ─────────────────────────────────────────────
# PAGE 2
# ─────────────────────────────────────────────
elif page == "🧠 Page 2: SQL Query Runner":
    st.header("🧠 SQL Query Runner")
    st.info("Select any predefined query and click **Run Query** to execute it against the SQLite database.")

    query_name = st.selectbox("🔽 Select Query", list(SQL_QUERIES.keys()))

    # Show SQL preview
    with st.expander("🔍 View SQL"):
        st.code(SQL_QUERIES[query_name], language="sql")

    if st.button("▶️ Run Query", type="primary"):
        conn = get_conn()
        try:
            result = pd.read_sql(SQL_QUERIES[query_name], conn)
            st.success(f"✅ Returned {len(result)} row(s)")
            st.dataframe(result, use_container_width=True)

            # Auto line chart if date column present
            if "date" in result.columns and len(result) > 5:
                num_cols = result.select_dtypes(include='number').columns.tolist()
                if num_cols:
                    st.line_chart(result.set_index("date")[num_cols])
        except Exception as e:
            st.error(f"Query error: {e}")
        conn.close()

# ─────────────────────────────────────────────
# PAGE 3
# ─────────────────────────────────────────────
elif page == "🪙 Page 3: Top 3 Crypto Analysis":
    st.header("🪙 Top 3 Cryptocurrency Analysis")

    conn = get_conn()
    try:
        coins = pd.read_sql("SELECT DISTINCT coin_id FROM crypto_prices", conn)["coin_id"].tolist()
    except:
        coins = ["bitcoin", "ethereum", "tether"]

    col1, col2, col3 = st.columns(3)
    with col1:
        selected_coin = st.selectbox("🪙 Select Coin", coins)
    with col2:
        start_date = st.date_input("📅 Start Date", value=datetime(2025, 1, 1), key="p3s")
    with col3:
        end_date = st.date_input("📅 End Date", value=datetime(2026, 6, 9), key="p3e")

    s, e = str(start_date), str(end_date)

    try:
        df_coin = pd.read_sql(f"""
            SELECT date, ROUND(price_usd, 2) as price_usd
            FROM crypto_prices
            WHERE coin_id='{selected_coin}' AND date BETWEEN '{s}' AND '{e}'
            ORDER BY date
        """, conn)

        if not df_coin.empty:
            # KPIs
            c1, c2, c3 = st.columns(3)
            c1.metric("📈 Max Price (USD)", f"${df_coin['price_usd'].max():,.2f}")
            c2.metric("📉 Min Price (USD)", f"${df_coin['price_usd'].min():,.2f}")
            c3.metric("📊 Avg Price (USD)", f"${df_coin['price_usd'].mean():,.2f}")

            st.subheader(f"📈 {selected_coin.capitalize()} Daily Price Trend")
            st.line_chart(df_coin.set_index("date")["price_usd"])

            st.subheader("📋 Daily Price Table")
            st.dataframe(df_coin, use_container_width=True, height=400)
        else:
            st.info("No data for selected date range.")
    except Exception as ex:
        st.error(f"Error: {ex}")

    conn.close()
```
