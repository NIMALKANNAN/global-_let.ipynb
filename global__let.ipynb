```python

df_literacy_filtered['youth_literacy_gender_gap'] = df_literacy_filtered['youth_literacy_rate_male'] - df_literacy_filtered['youth_literacy_rate_female']


df_literacy_filtered['youth_to_adult_ratio'] = df_literacy_filtered['youth_literacy_rate_male'].add(df_literacy_filtered['youth_literacy_rate_female']).div(2) / df_literacy_filtered['adult_literacy_rate']


df_literacy_filtered['youth_literacy_avg'] = (df_literacy_filtered['youth_literacy_rate_male'] + df_literacy_filtered['youth_literacy_rate_female']) / 2
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

 Univariate Analysis

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))


axes[0,0].hist(df['adult_literacy_rate'].dropna(), bins=20, color='steelblue', edgecolor='white')
axes[0,0].set_title('Adult Literacy Rate Distribution')
axes[0,0].set_xlabel('Rate (%)')


df[['youth_literacy_rate_male','youth_literacy_rate_female']].boxplot(ax=axes[0,1])
axes[0,1].set_title('Youth Literacy: Male vs Female')


gap = df.groupby('owid_region')['youth_literacy_gender_gap'].mean().sort_values()
gap.plot(kind='barh', ax=axes[0,2], color='coral')
axes[0,2].set_title('Avg Gender Gap by Region')


axes[1,0].hist(df['youth_literacy_avg'].dropna(), bins=20, color='mediumseagreen', edgecolor='white')
axes[1,0].set_title('Youth Avg Literacy Distribution')


df['year'].value_counts().sort_index().plot(kind='bar', ax=axes[1,1], color='mediumpurple')
axes[1,1].set_title('Observations per Year')


df['owid_region'].value_counts().plot(kind='pie', ax=axes[1,2], autopct='%1.0f%%')
axes[1,2].set_title('Records by Region')

plt.tight_layout()
plt.savefig('univariate.png', dpi=150)
plt.show()
```

 Bivariate Analysis

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))


for region, grp in df.groupby('owid_region'):
    axes[0,0].scatter(grp['adult_literacy_rate'], grp['youth_literacy_avg'],
                      label=region, alpha=0.7, s=40)
axes[0,0].set_title('Adult vs Youth Literacy')
axes[0,0].set_xlabel('Adult Literacy (%)')
axes[0,0].legend(fontsize=7)


num_cols = ['adult_literacy_rate','youth_literacy_rate_male',
            'youth_literacy_rate_female','youth_literacy_gender_gap',
            'youth_to_adult_ratio','youth_literacy_avg']
sns.heatmap(df[num_cols].corr(), annot=True, fmt='.2f',
            cmap='coolwarm', ax=axes[0,1])
axes[0,1].set_title('Correlation Heatmap')


axes[0,2].scatter(df['adult_literacy_rate'], df['youth_literacy_gender_gap'], alpha=0.5, color='salmon')
axes[0,2].set_title('Adult Literacy vs Gender Gap')


time_region = df.groupby(['year','owid_region'])['adult_literacy_rate'].mean().reset_index()
for region in df['owid_region'].unique():
    sub = time_region[time_region['owid_region']==region]
    axes[1,0].plot(sub['year'], sub['adult_literacy_rate'], marker='o', label=region)
axes[1,0].set_title('Adult Literacy Trend Over Time')
axes[1,0].legend(fontsize=7)


for region, grp in df.groupby('owid_region'):
    axes[1,1].scatter(grp['youth_literacy_rate_male'],
                      grp['youth_literacy_rate_female'], alpha=0.6, s=30, label=region)
axes[1,1].plot([0,100],[0,100],'k--', alpha=0.4, label='Equality')
axes[1,1].set_title('Male vs Female Youth Literacy')
axes[1,1].legend(fontsize=7)


reg = df.groupby('owid_region')[['youth_literacy_rate_male','youth_literacy_rate_female']].mean()
reg.plot(kind='bar', ax=axes[1,2])
axes[1,2].set_title('Regional Youth Literacy by Gender')
axes[1,2].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.savefig('bivariate.png', dpi=150)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Illiterate_filtered.csv")

palette = {"Low":"#D85A30", "Medium":"#EF9F27", "High":"#1D9E75"}

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="literacy_rate", y="illiteracy_rate",
                hue="literacy_category", palette=palette, alpha=0.7, s=60)
plt.title("Literacy Rate vs Illiteracy Rate")
plt.xlabel("Literacy Rate (%)")
plt.ylabel("Illiteracy Rate (%)")
plt.legend(title="Category")
plt.tight_layout()
plt.show()

num_cols = ["literacy_rate", "illiteracy_rate",
            "literacy_illiteracy_ratio", "years_since_2000"]

plt.figure(figsize=(6, 5))
sns.heatmap(df[num_cols].corr(), annot=True, fmt=".2f",
            cmap="coolwarm", center=0, linewidths=0.5)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()

time_df = df.groupby("years_since_2000")[["literacy_rate","illiteracy_rate"]].mean().reset_index()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(time_df["years_since_2000"], time_df["literacy_rate"],
             marker="o", color="steelblue")
axes[0].set_title("Avg Literacy Rate Over Time")
axes[0].set_xlabel("Years since 2000")
axes[0].set_ylabel("%")

axes[1].plot(time_df["years_since_2000"], time_df["illiteracy_rate"],
             marker="o", color="#D85A30")
axes[1].set_title("Avg Illiteracy Rate Over Time")
axes[1].set_xlabel("Years since 2000")

plt.tight_layout()
plt.show()

latest = df.sort_values("year").groupby("entity").last().reset_index()
top12 = latest.nlargest(12, "literacy_rate")

plt.figure(figsize=(8, 5))
sns.barplot(data=top12, y="entity", x="literacy_rate",
            palette="Greens_r")
plt.title("Top 12 Countries by Latest Literacy Rate")
plt.xlabel("Literacy Rate (%)")
plt.ylabel("")
plt.xlim(85, 100)
plt.tight_layout()
plt.show()
```

SQL Queries

```python
queries = {}
```

Top 5 countries with highest adult literacy in 2020

```python
queries["1. Top 5 countries with highest adult literacy in 2020"] = """
SELECT entity, adult_literacy_rate
FROM literacy_filtered
WHERE year = 2020
ORDER BY adult_literacy_rate DESC
LIMIT 5;"""
```

Find countries where female youth literacy < 80%.

```python
queries["2. Find countries where female youth literacy < 80%"] = """
SELECT DISTINCT entity, youth_literacy_rate_female
FROM literacy_filtered
WHERE youth_literacy_rate_female < 80
ORDER BY youth_literacy_rate_female DESC;"""
```

Average adult literacy per continent (owid region).



```python
queries["3. Average adult literacy per continent (owid region)"] = """
SELECT owid_region, ROUND(AVG(adult_literacy_rate), 2) as average_adult_literacy
FROM literacy_filtered
WHERE owid_region IS NOT NULL
GROUP BY owid_region
ORDER BY average_adult_literacy DESC;"""
```

4. Countries with illiteracy % > 20% in 2000.

```python
queries["4. Countries with illiteracy % > 20% in 2000"] = """
SELECT entity, illiteracy_rate
FROM Illiterate_filtered
WHERE year = 2000 AND illiteracy_rate > 20
ORDER BY illiteracy_rate DESC;"""
```

5. Trend of illiteracy % for India (2000–2020).

```python
queries["5. Trend of illiteracy % for India (2000-2020)"] = """
SELECT year, illiteracy_rate
FROM Illiterate_filtered
WHERE entity = 'India' AND year BETWEEN 2000 AND 2020
ORDER BY year;"""
```

6. Top 10 countries with largest illiterate population in the last year.

```python
queries["6. Top 10 countries with largest illiterate population in the last year"] = """
SELECT
    i.entity,
    ROUND((i.illiteracy_rate * g.population_historical / 100), 0) AS illiterate_population
FROM Illiterate_filtered i
JOIN gdp_average_filtered g ON i.entity = g.entity AND i.year = g.year
WHERE i.year = (SELECT MAX(year) FROM Illiterate_filtered)
  AND g.population_historical IS NOT NULL
ORDER BY illiterate_population DESC
LIMIT 10;"""
```

7. Find countries with avg_years_schooling > 7 and gdp_per_capita < 5000.


```python
queries = {}
queries["7. Countries with avg_years_schooling > 7 and gdp_per_capita < 5000"] = """
SELECT entity, gdp_per_capita, avg_years_of_education
FROM gdp_average_filtered
WHERE avg_years_of_education > 7 AND gdp_per_capita < 5000
ORDER BY gdp_per_capita DESC;"""
```

8.Rank countries by GDP per schooling for the year 2020.


```python
queries["8. Rank countries by GDP per schooling for 2020"] = """
SELECT entity, gdp_per_capita, avg_years_of_education,
       ROUND(gdp_per_capita / avg_years_of_education, 2) AS gdp_per_schooling
FROM gdp_average_filtered
WHERE year = 2020 AND avg_years_of_education IS NOT NULL AND avg_years_of_education > 0
ORDER BY gdp_per_schooling DESC;"""
```

9.Find global average schooling years per year.

```python
queries["9. Global average schooling years per year"] = """
SELECT year, ROUND(AVG(avg_years_of_education), 2) AS average_schooling_years
FROM gdp_average_filtered
WHERE avg_years_of_education IS NOT NULL
GROUP BY year
ORDER BY year;"""
```

10.   10. List top 10 countries in 2020 with highest GDP per capita but lowest average years of schooling(less than 6).
  11. Show countries where the illiterate population is high despite having more than 10 average years of schooling.
   12. Compare literacy rates and GDP per capita growth for a selected country over the last 20 years. (country of your choice)
  13. Show the difference between youth literacy male and female rates for countries with GDP per capita above $30,000 in 2020.


```python
queries["10. Top 10 countries (2020) by GDP per capita and low schooling"] = """
SELECT entity, gdp_per_capita, avg_years_of_education
FROM gdp_average_filtered
WHERE year = 2020 AND avg_per_capita IS NOT NULL AND avg_years_of_education < 6 AND avg_years_of_education IS NOT NULL
ORDER BY gdp_per_capita DESC
LIMIT 10;"""
```

```python
queries["11. Countries with high illiterate population despite high schooling"] = """
SELECT i.entity, i.year,
       ROUND((i.illiteracy_rate * g.population_historical / 100), 0) AS illiterate_population,
       g.avg_years_of_education
FROM Illiterate_filtered i
JOIN gdp_average_filtered g ON i.entity = g.entity AND i.year = g.year
WHERE g.avg_years_of_education > 10 AND i.illiteracy_rate IS NOT NULL AND g.population_historical IS NOT NULL
ORDER BY illiterate_population DESC
LIMIT 10;"""
```

```python
queries["12. Literacy vs GDP growth for India (last 20 years)"] = """
SELECT i.year, i.literacy_rate, g.gdp_growth_rate
FROM Illiterate_filtered i
JOIN gdp_average_filtered g ON i.entity = g.entity AND i.year = g.year
WHERE i.entity = 'India' AND i.year BETWEEN (SELECT MAX(year) FROM gdp_average_filtered) - 19 AND (SELECT MAX(year) FROM gdp_average_filtered)
ORDER BY i.year;"""
```

```python
queries["12. Literacy vs GDP growth for India (last 20 years)"] = """
SELECT i.year, i.literacy_rate, g.gdp_growth_rate
FROM Illiterate_filtered i
JOIN gdp_average_filtered g ON i.entity = g.entity AND i.year = g.year
WHERE i.entity = 'India' AND i.year BETWEEN (SELECT MAX(year) FROM gdp_average_filtered) - 19 AND (SELECT MAX(year) FROM gdp_average_filtered)
ORDER BY i.year;"""
```

```python
queries["13. Youth literacy gender gap for high GDP countries (2020)"] = """
SELECT l.entity, l.youth_literacy_rate_male, l.youth_literacy_rate_female,
       ROUND(l.youth_literacy_rate_male - l.youth_literacy_rate_female, 2) AS youth_gender_gap,
       g.gdp_per_capita
FROM literacy_filtered l
JOIN gdp_average_filtered g ON l.entity = g.entity AND l.year = g.year
WHERE l.year = 2020 AND g.gdp_per_capita > 30000
  AND l.youth_literacy_rate_male IS NOT NULL AND l.youth_literacy_rate_female IS NOT NULL
ORDER BY youth_gender_gap DESC;"""
```

```python
%%writefile literate.py

import streamlit as st
import pandas as pd
import sqlite3
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import base64
from io import BytesIO

# Helper function to convert matplotlib figures to HTML
def generate_plot_html(fig):
    buf = BytesIO()
    fig.savefig(buf, format="png", bbox_inches="tight")
    data = base64.b64encode(buf.getvalue()).decode("utf-8")
    buf.close()
    plt.close(fig)
    return f"<img src='data:image/png;base64,{data}' style='width:100%;'>"

# Define SQL_QUERIES dictionary
SQL_QUERIES = {
    "1. Top 5 countries with highest adult literacy in 2020": """
SELECT entity, adult_literacy_rate
FROM literacy_filtered
WHERE year = 2020
ORDER BY adult_literacy_rate DESC
LIMIT 5;""",
    "2. Find countries where female youth literacy < 80%": """
SELECT DISTINCT entity, youth_literacy_rate_female
FROM literacy_filtered
WHERE youth_literacy_rate_female < 80
ORDER BY youth_literacy_rate_female DESC;""",
    "3. Average adult literacy per continent (owid region)": """
SELECT owid_region, ROUND(AVG(adult_literacy_rate), 2) as average_adult_literacy
FROM literacy_filtered
WHERE owid_region IS NOT NULL
GROUP BY owid_region
ORDER BY average_adult_literacy DESC;""",
    "4. Countries with illiteracy % > 20% in 2000": """
SELECT entity, illiteracy_rate
FROM Illiterate_filtered
WHERE year = 2000 AND illiteracy_rate > 20
ORDER BY illiteracy_rate DESC;""",
    "5. Trend of illiteracy % for India (2000-2020)": """
SELECT year, illiteracy_rate
FROM Illiterate_filtered
WHERE entity = 'India' AND year BETWEEN 2000 AND 2020
ORDER BY year;""",
    "6. Top 10 countries with largest illiterate population in the last year": """
SELECT
    i.entity,
    ROUND((i.illiteracy_rate * g.population_historical / 100), 0) AS illiterate_population
FROM Illiterate_filtered i
JOIN gdp_average_filtered g ON i.entity = g.entity AND i.year = g.year
WHERE i.year = (SELECT MAX(year) FROM Illiterate_filtered)
  AND g.population_historical IS NOT NULL
ORDER BY illiterate_population DESC
LIMIT 10;""",
    "7. Countries with avg_years_schooling > 7 and gdp_per_capita < 5000": """
SELECT entity, gdp_per_capita, avg_years_of_education
FROM gdp_average_filtered
WHERE avg_years_of_education > 7 AND gdp_per_capita < 5000
ORDER BY gdp_per_capita DESC;""",
    "8. Rank countries by GDP per schooling for 2020": """
SELECT entity, gdp_per_capita, avg_years_of_education,
       ROUND(gdp_per_capita / avg_years_of_education, 2) AS gdp_per_schooling
FROM gdp_average_filtered
WHERE year = 2020 AND avg_years_of_education IS NOT NULL AND avg_years_of_education > 0
ORDER BY gdp_per_schooling DESC;""",
    "9. Global average schooling years per year": """
SELECT year, ROUND(AVG(avg_years_of_education), 2) AS average_schooling_years
FROM gdp_average_filtered
WHERE avg_years_of_education IS NOT NULL
GROUP BY year
ORDER BY year;""",
    "10. Top 10 countries (2020) by GDP per capita and low schooling": """
SELECT entity, gdp_per_capita, avg_years_of_education
FROM gdp_average_filtered
WHERE year = 2020 AND gdp_per_capita IS NOT NULL AND avg_years_of_education < 6 AND avg_years_of_education IS NOT NULL
ORDER BY gdp_per_capita DESC
LIMIT 10;""",
    "11. Countries with high illiterate population despite high schooling": """
SELECT i.entity, i.year,
       ROUND((i.illiteracy_rate * g.population_historical / 100), 0) AS illiterate_population,
       g.avg_years_of_education
FROM Illiterate_filtered i
JOIN gdp_average_filtered g ON i.entity = g.entity AND i.year = g.year
WHERE g.avg_years_of_education > 10 AND i.illiteracy_rate IS NOT NULL AND g.population_historical IS NOT NULL
ORDER BY illiterate_population DESC
LIMIT 10;""",
    "12. Literacy vs GDP growth for India (last 20 years)": """
SELECT i.year, i.literacy_rate, g.gdp_growth_rate
FROM Illiterate_filtered i
JOIN gdp_average_filtered g ON i.entity = g.entity AND i.year = g.year
WHERE i.entity = 'India' AND i.year BETWEEN (SELECT MAX(year) FROM gdp_average_filtered) - 19 AND (SELECT MAX(year) FROM gdp_average_filtered)
ORDER BY i.year;""",
    "13. Youth literacy gender gap for high GDP countries (2020)": """
SELECT l.entity, l.youth_literacy_rate_male, l.youth_literacy_rate_female,
       ROUND(l.youth_literacy_rate_male - l.youth_literacy_rate_female, 2) AS youth_gender_gap,
       g.gdp_per_capita
FROM literacy_filtered l
JOIN gdp_average_filtered g ON l.entity = g.entity AND l.year = g.year
WHERE l.year = 2020 AND g.gdp_per_capita > 30000
  AND l.youth_literacy_rate_male IS NOT NULL AND l.youth_literacy_rate_female IS NOT NULL
ORDER BY youth_gender_gap DESC;"""
}

def get_conn():
    return sqlite3.connect("mydb.db")

st.set_page_config(
    page_title="Literacy & GDP Explorer",
    page_icon="📚",
    layout="wide",
    initial_sidebar_state="expanded",
)

st.sidebar.title("📚 Literacy & GDP Explorer")
st.sidebar.markdown("---")
st.sidebar.markdown(
    """
    **Navigate the pages:**
    - 🗄️ SQL Query Executor
    - 📊 EDA Visualizations
    - 🌍 Country Profile
    """
)

st.title("Welcome to the Literacy & GDP Explorer")
st.markdown(
    """
    This multi-page Streamlit app lets you explore global **literacy rates**
    and **GDP per capita** data across countries and years.

    ### Pages
    | Page | Description |
    |------|-------------|
    | 🗄️ **SQL Query Executor** | Write SQL against the datasets; view results as tables and charts |
    | 📊 **EDA Visualizations** | Explore pre-built Matplotlib / Seaborn / Plotly charts |
    | 🌍 **Country Profile** | Pick any country and see all indicators over time |

    Use the **sidebar** to navigate.
    """
)

# Page navigation
selected_page = st.sidebar.radio("Go to", ["Home", "🗄️ SQL Query Executor", "📊 EDA Visualizations", "🌍 Country Profile"])

if selected_page == "Home":
    st.subheader("Overview")
    st.write("Select a page from the sidebar to begin your exploration.")

elif selected_page == "🗄️ SQL Query Executor":
    st.header("🗄️ SQL Query Executor")
    st.write("Select a predefined SQL query to explore the data.")

    # Dropdown to select queries
    query_name = st.selectbox("Select a Query", list(SQL_QUERIES.keys()))

    if query_name:
        st.code(SQL_QUERIES[query_name], language="sql")

        if st.button("Run Query"):
            conn = get_conn()
            try:
                df_result = pd.read_sql(SQL_QUERIES[query_name], conn)
                st.success("Query executed successfully!")
                st.dataframe(df_result)

                # Attempt to plot if 'year' or 'date' column is present and there are numeric columns
                date_cols = ['year', 'date']
                found_date_col = None
                for col in date_cols:
                    if col in df_result.columns:
                        found_date_col = col
                        break

                if found_date_col:
                    numeric_cols = df_result.select_dtypes(include=['number']).columns.tolist()
                    if numeric_cols and len(df_result) > 1:
                        # Exclude date/year from numeric columns for plotting y-axis
                        plot_cols = [c for c in numeric_cols if c != found_date_col]
                        if plot_cols:
                            st.subheader("Visualization of Query Result")
                            st.line_chart(df_result.set_index(found_date_col)[plot_cols])
                        else:
                            st.info("No numeric columns available for plotting.")
                    else:
                        st.info("Not enough data or no numeric columns for plotting.")
                else:
                    st.info("No 'year' or 'date' column found to create a time-series plot.")

            except Exception as e:
                st.error(f"Error executing query: {e}")
            finally:
                conn.close()

elif selected_page == "📊 EDA Visualizations":
    st.header("📊 EDA Visualizations")
    st.write("Explore key insights and patterns through various visualizations.")

    conn = get_conn()

    try:
        # Load filtered dataframes from SQLite
        df_literacy = pd.read_sql("SELECT * FROM literacy_filtered", conn)
        df_illiterate = pd.read_sql("SELECT * FROM Illiterate_filtered", conn)
        df_gdp_average = pd.read_sql("SELECT * FROM gdp_average_filtered", conn)

        # --- Literacy Data Visualizations ---
        st.subheader("📚 Literacy Data Analysis")

        # Univariate Analysis for df_literacy
        st.markdown("##### Univariate Analysis (Literacy)")
        fig, axes = plt.subplots(2, 3, figsize=(18, 10))

        axes[0,0].hist(df_literacy['adult_literacy_rate'].dropna(), bins=20, color='steelblue', edgecolor='white')
        axes[0,0].set_title('Adult Literacy Rate Distribution')
        axes[0,0].set_xlabel('Rate (%)')

        df_literacy[['youth_literacy_rate_male','youth_literacy_rate_female']].boxplot(ax=axes[0,1])
        axes[0,1].set_title('Youth Literacy: Male vs Female')

        gap = df_literacy.groupby('owid_region')['youth_literacy_gender_gap'].mean().sort_values()
        gap.plot(kind='barh', ax=axes[0,2], color='coral')
        axes[0,2].set_title('Avg Gender Gap by Region')

        axes[1,0].hist(df_literacy['youth_literacy_avg'].dropna(), bins=20, color='mediumseagreen', edgecolor='white')
        axes[1,0].set_title('Youth Avg Literacy Distribution')

        df_literacy['year'].value_counts().sort_index().plot(kind='bar', ax=axes[1,1], color='mediumpurple')
        axes[1,1].set_title('Observations per Year')

        df_literacy['owid_region'].value_counts().plot(kind='pie', ax=axes[1,2], autopct='%1.0f%%')
        axes[1,2].set_title('Records by Region')

        plt.tight_layout()
        st.markdown(generate_plot_html(fig), unsafe_allow_html=True)

        # Bivariate Analysis for df_literacy
        st.markdown("##### Bivariate Analysis (Literacy)")
        fig, axes = plt.subplots(2, 3, figsize=(18, 10))

        for region, grp in df_literacy.groupby('owid_region'):
            axes[0,0].scatter(grp['adult_literacy_rate'], grp['youth_literacy_avg'],
                              label=region, alpha=0.7, s=40)
        axes[0,0].set_title('Adult vs Youth Literacy')
        axes[0,0].set_xlabel('Adult Literacy (%)')
        axes[0,0].legend(fontsize=7)

        num_cols_literacy = ['adult_literacy_rate','youth_literacy_rate_male',
                    'youth_literacy_rate_female','youth_literacy_gender_gap',
                    'youth_to_adult_ratio','youth_literacy_avg']
        sns.heatmap(df_literacy[num_cols_literacy].corr(), annot=True, fmt='.2f',
                    cmap='coolwarm', ax=axes[0,1])
        axes[0,1].set_title('Correlation Heatmap')

        axes[0,2].scatter(df_literacy['adult_literacy_rate'], df_literacy['youth_literacy_gender_gap'], alpha=0.5, color='salmon')
        axes[0,2].set_title('Adult Literacy vs Gender Gap')

        time_region = df_literacy.groupby(['year','owid_region'])['adult_literacy_rate'].mean().reset_index()
        for region in df_literacy['owid_region'].unique():
            sub = time_region[time_region['owid_region']==region]
            axes[1,0].plot(sub['year'], sub['adult_literacy_rate'], marker='o', label=region)
        axes[1,0].set_title('Adult Literacy Trend Over Time')
        axes[1,0].legend(fontsize=7)

        for region, grp in df_literacy.groupby('owid_region'):
            axes[1,1].scatter(grp['youth_literacy_rate_male'],
                              grp['youth_literacy_rate_female'], alpha=0.6, s=30, label=region)
        axes[1,1].plot([0,100],[0,100],'k--', alpha=0.4, label='Equality')
        axes[1,1].set_title('Male vs Female Youth Literacy')
        axes[1,1].legend(fontsize=7)

        reg = df_literacy.groupby('owid_region')[['youth_literacy_rate_male','youth_literacy_rate_female']].mean()
        reg.plot(kind='bar', ax=axes[1,2])
        axes[1,2].set_title('Regional Youth Literacy by Gender')
        axes[1,2].tick_params(axis='x', rotation=30)

        plt.tight_layout()
        st.markdown(generate_plot_html(fig), unsafe_allow_html=True)

        # --- Illiteracy Data Visualizations ---
        st.subheader("📉 Illiteracy Data Analysis")

        # Univariate Analysis for df_illiterate
        st.markdown("##### Univariate Analysis (Illiteracy)")
        fig = plt.figure(figsize=(7, 4))
        sns.histplot(df_illiterate["literacy_rate"], bins=15, kde=True, color="steelblue")
        plt.title("Distribution of Literacy Rate")
        plt.xlabel("Literacy Rate (%)")
        plt.ylabel("Count")
        plt.tight_layout()
        st.markdown(generate_plot_html(fig), unsafe_allow_html=True)

        fig = plt.figure(figsize=(7, 4))
        order = ["Low", "Medium", "High"]
        sns.boxplot(data=df_illiterate, x="literacy_category", y="literacy_rate",
                    order=order, palette={"Low":"#D85A30","Medium":"#EF9F27","High":"#1D9E75"})
        plt.title("Literacy Rate by Category")
        plt.tight_layout()
        st.markdown(generate_plot_html(fig), unsafe_allow_html=True)

        fig = plt.figure(figsize=(7, 4))
        cat_counts = df_illiterate["literacy_category"].value_counts().reindex(["Low","Medium","High"])
        cat_counts.plot(kind="bar", color=["#D85A30","#EF9F27","#1D9E75"], edgecolor="white")
        plt.title("Observation Count by Literacy Category")
        plt.ylabel("Count")
        plt.xticks(rotation=0)
        plt.tight_layout()
        st.markdown(generate_plot_html(fig), unsafe_allow_html=True)

        # Bivariate Analysis for df_illiterate
        st.markdown("##### Bivariate Analysis (Illiteracy)")
        palette_illiterate = {"Low":"#D85A30", "Medium":"#EF9F27", "High":"#1D9E75"}

        fig = plt.figure(figsize=(8, 5))
        sns.scatterplot(data=df_illiterate, x="literacy_rate", y="illiteracy_rate",
                        hue="literacy_category", palette=palette_illiterate, alpha=0.7, s=60)
        plt.title("Literacy Rate vs Illiteracy Rate")
        plt.xlabel("Literacy Rate (%)")
        plt.ylabel("Illiteracy Rate (%)")
        plt.legend(title="Category")
        plt.tight_layout()
        st.markdown(generate_plot_html(fig), unsafe_allow_html=True)

        num_cols_illiterate = ["literacy_rate", "illiteracy_rate",
                    "literacy_illiteracy_ratio", "years_since_2000"]

        fig = plt.figure(figsize=(6, 5))
        sns.heatmap(df_illiterate[num_cols_illiterate].corr(), annot=True, fmt=".2f",
                    cmap="coolwarm", center=0, linewidths=0.5)
        plt.title("Correlation Heatmap")
        plt.tight_layout()
        st.markdown(generate_plot_html(fig), unsafe_allow_html=True)

        time_df_illiterate = df_illiterate.groupby("years_since_2000")[["literacy_rate","illiteracy_rate"]].mean().reset_index()

        fig, axes = plt.subplots(1, 2, figsize=(12, 4))

        axes[0].plot(time_df_illiterate["years_since_2000"], time_df_illiterate["literacy_rate"],
                     marker="o", color="steelblue")
        axes[0].set_title("Avg Literacy Rate Over Time")
        axes[0].set_xlabel("Years since 2000")
        axes[0].set_ylabel("%")

        axes[1].plot(time_df_illiterate["years_since_2000"], time_df_illiterate["illiteracy_rate"],
                     marker="o", color="#D85A30")
        axes[1].set_title("Avg Illiteracy Rate Over Time")
        axes[1].set_xlabel("Years since 2000")

        plt.tight_layout()
        st.markdown(generate_plot_html(fig), unsafe_allow_html=True)

        latest_illiterate = df_illiterate.sort_values("year").groupby("entity").last().reset_index()
        top12_illiterate = latest_illiterate.nlargest(12, "literacy_rate")

        fig = plt.figure(figsize=(8, 5))
        sns.barplot(data=top12_illiterate, y="entity", x="literacy_rate",
                    palette="Greens_r")
        plt.title("Top 12 Countries by Latest Literacy Rate")
        plt.xlabel("Literacy Rate (%)")
        plt.ylabel("")
        plt.xlim(85, 100)
        plt.tight_layout()
        st.markdown(generate_plot_html(fig), unsafe_allow_html=True)

        # --- GDP & Education Data Visualizations ---
        st.subheader("💰 GDP and Education Data Analysis")

        # Univariate Analysis for df_gdp_average
        st.markdown("##### Univariate Analysis (GDP & Education)")
        fig = plt.figure(figsize=(8, 4))
        sns.histplot(df_gdp_average['gdp_per_capita'].dropna(), bins=40, kde=True, color='steelblue')
        plt.title('Distribution of GDP per capita')
        plt.xlabel('GDP per capita (USD)')
        plt.tight_layout()
        st.markdown(generate_plot_html(fig), unsafe_allow_html=True)

        fig = plt.figure(figsize=(10, 5))
        sns.boxplot(data=df_gdp_average, x='owid_region_x', y='avg_years_of_education', palette='Set2')
        plt.xticks(rotation=30)
        plt.title('Avg. years of education by region')
        plt.tight_layout()
        st.markdown(generate_plot_html(fig), unsafe_allow_html=True)

        fig = plt.figure(figsize=(8, 4))
        sns.histplot(df_gdp_average['literacy_rate'].dropna(), bins=20, kde=True, color='mediumseagreen')
        plt.title('Literacy rate distribution')
        plt.tight_layout()
        st.markdown(generate_plot_html(fig), unsafe_allow_html=True)

        # Bivariate Analysis for df_gdp_average
        st.markdown("##### Bivariate Analysis (GDP & Education)")

        fig = plt.figure(figsize=(7,6))
        sns.scatterplot(data=df_gdp_average, x='literacy_rate', y='gdp_per_capita',
                        hue='owid_region_x', alpha=0.6)
        plt.title('GDP per capita vs literacy rate')
        plt.yscale('log')
        plt.tight_layout()
        st.markdown(generate_plot_html(fig), unsafe_allow_html=True)

        num_cols_gdp = ['gdp_per_capita', 'literacy_rate', 'avg_years_of_education',
                    'gdp_growth_rate', 'pop_growth_rate']

        fig = plt.figure(figsize=(7, 6))
        corr_gdp = df_gdp_average[num_cols_gdp].corr()
        sns.heatmap(corr_gdp, annot=True, fmt='.2f', cmap='coolwarm',
                    center=0, square=True, linewidths=0.5)
        plt.title('Correlation heatmap')
        plt.tight_layout()
        st.markdown(generate_plot_html(fig), unsafe_allow_html=True)

        trend_gdp = df_gdp_average.groupby(['year', 'owid_region_x'])['gdp_per_capita'].mean().reset_index()

        fig = plt.figure(figsize=(10,5))
        sns.lineplot(data=trend_gdp, x='year', y='gdp_per_capita',
                     hue='owid_region_x', marker='o')
        plt.title('GDP per capita over time by region')
        plt.tight_layout()
        st.markdown(generate_plot_html(fig), unsafe_allow_html=True)

    except Exception as e:
        st.error(f"Error loading data or generating visualizations: {e}")
    finally:
        conn.close()

elif selected_page == "🌍 Country Profile":
    st.header("🌍 Country Profile")
    st.write("Explore various indicators for a selected country over time.")

    conn = get_conn()

    try:
        countries_literacy = pd.read_sql("SELECT DISTINCT entity FROM literacy_filtered ORDER BY entity", conn)['entity'].tolist()
        countries_illiterate = pd.read_sql("SELECT DISTINCT entity FROM Illiterate_filtered ORDER BY entity", conn)['entity'].tolist()
        countries_gdp = pd.read_sql("SELECT DISTINCT entity FROM gdp_average_filtered ORDER BY entity", conn)['entity'].tolist()
        all_countries = sorted(list(set(countries_literacy + countries_illiterate + countries_gdp)))
    except Exception as e:
        st.error(f"Error fetching country list: {e}")
        all_countries = []

    if not all_countries:
        st.warning("No country data available in the database.")
        conn.close()
        st.stop()

    selected_country = st.selectbox("Select a Country", all_countries)

    indicators_map = {
        "Adult Literacy Rate": {"table": "literacy_filtered", "column": "adult_literacy_rate"},
        "Youth Literacy Rate (Male)": {"table": "literacy_filtered", "column": "youth_literacy_rate_male"},
        "Youth Literacy Rate (Female)": {"table": "literacy_filtered", "column": "youth_literacy_rate_female"},
        "Youth Literacy Gender Gap": {"table": "literacy_filtered", "column": "youth_literacy_gender_gap"},
        "Youth to Adult Literacy Ratio": {"table": "literacy_filtered", "column": "youth_to_adult_ratio"},
        "Average Youth Literacy Rate": {"table": "literacy_filtered", "column": "youth_literacy_avg"},
        "Illiteracy Rate": {"table": "Illiterate_filtered", "column": "illiteracy_rate"},
        "Literacy Rate": {"table": "Illiterate_filtered", "column": "literacy_rate"},
        "Literacy-Illiteracy Ratio": {"table": "Illiterate_filtered", "column": "literacy_illiteracy_ratio"},
        "GDP per Capita": {"table": "gdp_average_filtered", "column": "gdp_per_capita"},
        "Average Years of Education": {"table": "gdp_average_filtered", "column": "avg_years_of_education"},
        "Population Historical": {"table": "gdp_average_filtered", "column": "population_historical"},
        "GDP Growth Rate": {"table": "gdp_average_filtered", "column": "gdp_growth_rate"},
        "Population Growth Rate": {"table": "gdp_average_filtered", "column": "pop_growth_rate"},
        "Education Change": {"table": "gdp_average_filtered", "column": "edu_change"}
    }
    indicator_names = list(indicators_map.keys())
    selected_indicator_name = st.selectbox("Select an Indicator", indicator_names)

    if selected_country and selected_indicator_name:
        indicator_info = indicators_map[selected_indicator_name]
        table_name = indicator_info["table"]
        column_name = indicator_info["column"]

        query = f"""
            SELECT year, "{column_name}"
            FROM "{table_name}"
            WHERE entity = '{selected_country}'
            ORDER BY year;
        """

        st.subheader(f"{selected_indicator_name} for {selected_country}")

        try:
            df_indicator = pd.read_sql(query, conn)
            if not df_indicator.empty:
                st.dataframe(df_indicator, use_container_width=True)

                st.subheader("Trend over Time")
                df_indicator_plot = df_indicator.set_index('year')
                st.line_chart(df_indicator_plot)
            else:
                st.info(f"No data available for {selected_indicator_name} in {selected_country}.")
        except Exception as e:
            st.error(f"Error fetching data for {selected_indicator_name}: {e}")

    conn.close()
```
